```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.arange(1, 31)
y = 2 * x
plt.xlabel("x")
plt.ylabel("y")
plt.plot(x, y)
plt.show()
```

```python
financial_data = pd.read_csv('fdata.csv')
print(financial_data)
```

```python
df = pd.DataFrame(financial_data)
df['Date'] = pd.to_datetime(df['Date'])
plt.plot(df['Date'], df['Open'], label="Open")
plt.plot(df['Date'], df['High'], label="High", linewidth=2, linestyle=":", color='yellow')
plt.plot(df['Date'], df['Low'], label="Low", linewidth=3, linestyle='-.')
plt.plot(df['Date'], df['Close'], label="Close", linewidth=5, linestyle='--')
plt.show()
```

```python
ry1 = np.linspace(20, 40, 5)
ry2 = np.linspace(40, 10, 5)
ry = np.hstack((ry1, ry2[1:]))
print(ry)

gy1 = np.linspace(40, 10, 5)
gy2 = np.linspace(10, 30, 5)
gy = np.hstack((gy1, gy2[1:]))
print(gy)
```

```python
px = np.linspace(10, 30, 9)
plt.plot(px, ry, linewidth=3, linestyle=':', color='r')
plt.plot(px, gy, linewidth=5, linestyle='--', color='g')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend(["red-dotted", "green-dashed"], loc=1)
plt.title("Plot with two or more lines with different style")
plt.show()
```

```python
prog_lang = ["Python", "Java", "C", "C++", "R", "JavaScript", "C#"]
popularity = [100, 96.3, 94.4, 87.5, 81.5, 79.4, 74.5]
plt.bar(prog_lang, popularity)
plt.show()
```

```python
plt.barh(prog_lang, popularity)
plt.show()
```

```python
colors = ['red', 'yellow', 'green', 'blue', 'orange', 'purple', 'grey']
plt.bar(prog_lang, popularity, color=colors)
plt.show()
```

```python
plt.bar(prog_lang, popularity, color=colors)
for x, y in zip(prog_lang, popularity):
    plt.text(x, y, '%.2f'%y, ha='center')
plt.show()
```

```python
plt.pie(popularity, labels=prog_lang, autopct='%.2f%%')
plt.show()
```

```python
explode = [0.3, 0.5, 1, 0, 0, 0, 0]
plt.pie(popularity, labels=prog_lang, autopct='%.2f%%', explode=explode)
plt.show()
```
