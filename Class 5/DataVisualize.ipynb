```python
x = np.arange(1, 31)
y1 = x * 2
y2 = np.sin(np.pi + x) * 30 + 30
plt.figure(figsize=(6, 3))
plt.plot(x, y1, linewidth=5, color='r', linestyle='--')
plt.plot(x, y2, linewidth=2, color='g', linestyle=':')
plt.xlabel("x values")
plt.ylabel("y values")
plt.title("Task 3")
plt.legend(['straight', 'sin'])
plt.show()
```

Task 4

```python
ry1 = np.linspace(20, 40, 5)
ry2 = np.linspace(40, 10, 5)
ry = np.hstack((ry1, ry2[1:]))
print(ry)

gy1 = np.linspace(40, 10, 5)
gy2 = np.linspace(10, 30, 5)
gy = np.hstack((gy1, gy2[1:]))
print(gy)
```

```python
x = np.linspace(10, 30, 9)
plt.figure(figsize=(6, 3))
plt.plot(x, ry, linewidth=2, color='r', linestyle=':')
plt.plot(x, gy, linewidth=4, color='g', linestyle='--')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Plot with two or more lines with different styles")
plt.legend(['red-dotted', 'green-dashed'], loc=10)
plt.show()
```

Task 6

```python
x = np.arange(0, 5, 0.1)
y1 = np.sin(2 * np.pi * x)
y2 = np.cos(2 * np.pi * x)
y3 = np.tan(2 * np.pi * x)
plt.subplot(311)
plt.plot(x, y1, linestyle='-.', color='r')
plt.subplot(312)
plt.plot(x, y2, linestyle='--', color='g')
plt.subplot(313)
plt.plot(x, y3, linestyle=':', color='b')
plt.show()
```

Task 7

```python
index = ['Python', 'Java', 'C', 'C++', 'R', 'Javascript', 'C#']
values = [100, 96.3, 94.4, 87.5, 81.5, 79.4, 74.5]
plt.figure(figsize=(6, 4))
plt.bar(index, values)
plt.xlabel("Programming Languages")
plt.ylabel("Popularity")
plt.title("Popularity of Programming")
plt.show()
```

Task 8

```python
plt.figure(figsize=(6, 4))
plt.barh(index, values)
plt.ylabel("Programming Languages")
plt.xlabel("Popularity")
plt.title("Popularity of Programming")
plt.show()
```

Task 9

```python
colors = ['red', 'green', 'blue', 'yellow', 'orange', 'gray', 'purple']
plt.figure(figsize=(6, 4))
plt.bar(index, values, color=colors)
plt.xlabel("Programming Languages")
plt.ylabel("Popularity")
plt.title("Popularity of Programming")
plt.show()
```

Task 10

```python
plt.figure(figsize=(6, 4))
plt.bar(index, values)
for x, y in zip(index, values):
    plt.text(x, y, '%.2f'%y, ha='center')
plt.xlabel("Programming Languages")
plt.ylabel("Popularity")
plt.title("Popularity of Programming")
plt.show()
```

Task 11

```python
plt.figure(figsize=(6, 4))
plt.pie(values, labels=index, autopct='%.2f%%')
plt.title("Popularity of Programming")
plt.show()
```

Task 12

```python
explode = [0.3, 0.1, 0.1, 0, 0, 0, 0]
plt.figure(figsize=(6, 4))
plt.pie(values, labels=index, autopct='%.2f%%', explode=explode, startangle=45)
plt.title("Popularity of Programming")
plt.show()
```
